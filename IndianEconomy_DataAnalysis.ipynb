```python
# Giving Size
plt.figure(figsize=(10,5))

# Title
plt.title("Net Exports Of Goods And Services Of India From 1960 To 2020")
plt.xlabel("Year",fontdict={'fontweight':'bold', 'fontsize':15})
plt.ylabel("Exports And Imports",fontdict={'fontweight':'bold', 'fontsize':15})

# Plotting The Values
plt.plot(Net_Export.Year, Net_Export['Exports of goods and services (% of GDP)'], label='Exports', color='blue', marker='.',markersize=5,markeredgecolor='black')
plt.plot(Net_Export.Year, Net_Export['Imports of goods and services (% of GDP)'], label='Imports', color='green', marker='.',markersize=5,markeredgecolor='black')
plt.plot(Net_Export.Year, Net_Export['Net Exports'], label='NetExports', color='red', marker='.',markersize=5,markeredgecolor='black')

# Changing The X And Y Values Slightly
plt.xticks(Net_Export.Year[::4])
ytick_locations = range(-6, 36, 4)
plt.yticks(ytick_locations)

plt.legend()
plt.show()
```

```python
# Giving Size
plt.figure(figsize=(10,5))

# Title
plt.title("Exports Of Goods And Services Of India From 1960 To 2020")
plt.xlabel("Year",fontdict={'fontweight':'bold', 'fontsize':15})
plt.ylabel("Exports ",fontdict={'fontweight':'bold', 'fontsize':15})

# Plotting The Values
plt.plot(Net_Export.Year, Net_Export['Exports of goods and services (% of GDP)'], label='Exports', color='blue', marker='.',markersize=5,markeredgecolor='black')


# Changing The X And Y Values Slightly
plt.xticks(Net_Export.Year[::4])
ytick_locations = range(-6, 36, 4)
plt.yticks(ytick_locations)

plt.legend()
plt.show()
```
