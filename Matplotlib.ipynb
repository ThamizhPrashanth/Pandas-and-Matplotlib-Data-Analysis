```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

**BASIC GRAPH**

```python
x=[1,2,3,4,5]
y=[0,2,4,6,8]
#plt.plot([1,2,3],[2,4,6])
# plt.plot(x,y)
# plt.show()

# Resize The Graph
plt.figure(figsize=(5,2),dpi=100)


plt.plot(x,y, label='2x', color='red', linewidth=2, marker='.', linestyle='--', markersize=10, markeredgecolor='black') # Giving Values

## Line Number Two
# x2 = (0,4.5,0.5,0.9)
# y2 = (1,3,5,7.5)
# plt.plot(x2, y2, label='x2', color='blue', linewidth=2, marker='.', linestyle='--', markersize=10, markeredgecolor='black')
x2=np.arange(0,4.5,0.5)
plt.plot(x2[:6], x2[:6]**2,label='x2', color='blue', linewidth=2, marker='.', linestyle='--', markersize=10, markeredgecolor='black')
plt.plot(x2[5:], x2[5:]**2,label='x^2', color='green', linewidth=2, marker='.', linestyle='--', markersize=10, markeredgecolor='black')


plt.title('Sample Graph', fontdict={'fontsize': 20})  # To Give a Title
plt.xlabel('X Axis')  # To Give Label To x-axis
plt.ylabel('Y Axis')  # To Give Label To y-axis


# 'fontname': 'Comic Sans MS'

# Change int into Whole Num
plt.xticks([0,1,2,3,4,5])
plt.yticks([2,4,6,8,10])

# To Add Legend
plt.legend()

plt.savefig('BasicGraph.png', dpi=150)

# Show Plot
plt.show()
```

**Bar Chart**

```python
labels = ['A','B','C','D']
values = [1,5,9,4]

bars = plt.bar(labels, values)

# To Make Hatch On The Bar
bars[0].set_hatch('O')
bars[1].set_hatch('-')
bars[2].set_hatch('*')
bars[3].set_hatch('/')

plt.figure(figsize=(6,4))
plt.show()
```

**HISTOGRAM**

```python
bins=[40,50,60,70,80,90,100]
plt.hist(fifa.Overall, bins=bins, color='black')

plt.xticks(bins)
plt.ylabel('Number Of Players')
plt.xlabel('Skill Level')
plt.title('Distribution of Player Skills In FIFA')

plt.show()
```

**PIE CHART 2**

```python
fifa.Weight
```

```python
fifa.Weight = [int( x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]
plt.style.use('ggplot')
```

```python
fifa.Weight[0]
```

```python
lean = fifa.loc[fifa.Weight < 125].count()[0]
normal = fifa.loc[(fifa.Weight) >= 125 & (fifa.Weight < 150)].count()[0]
fit = fifa.loc[(fifa.Weight) >= 150 & (fifa.Weight < 175)].count()[0]
medium_obese = fifa.loc[(fifa.Weight) >= 175 & (fifa.Weight < 200)].count()[0]
obese = fifa.loc[fifa.Weight >= 200].count()[0]
```

```python
Weights =[lean,normal,fit,medium_obese,obese]
labels = ['Under 125', '125-150', '150-175','175-200', 'Over 200']
explode = (.4,.2,0,0,.4)
```

```python
plt.pie(Weights, labels=labels, autopct = '%.2f %%', pctdistance= .8, explode=explode)

plt.title('Weight Distribution  Of FIFA Players(Lbs)')
plt.show()
```

**BOX & WHISKER PLOT**

```python
plt.style.use('default')
plt.figure(figsize=(5,8))
barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
revs = fifa.loc[fifa.Club == 'New England Revolution']['Overall']

labels = ['FC Barcelona', 'Real Madrid', 'NE Revolution']
#color = ['red','gold','black']
plt.title('Professional Football Team Comparison')
plt.ylabel('FIFA Overall Rating')

plt.boxplot([barcelona, madrid, revs], labels=labels)
plt.show()
```
